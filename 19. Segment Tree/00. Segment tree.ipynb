```python
# Lets build a Node
class TreeNode:
    def __init__(self,left=None,right=None,val=None,left_child=None,right_child=None):
        self.val   = val
        self.left  = left
        self.right = right
        self.left_child = left_child
        self.right_child = right_child
```

```python
class Sum:
    def __init__(self):
        self.root = None
        self.operation = None
        
    def _buildTree(self, arr:list, left:int, right:int) -> None:
        if left > right: return None
        if left == right: return TreeNode(left, right, arr[left])
        mid = (left + right) // 2
        
        root = TreeNode(left, right)
        root.left_child  = self._buildTree(arr, left, mid)
        root.right_child = self._buildTree(arr, mid+1, right)
        root.val   = eval(self.operation)([root.left_child.val, root.right_child.val])
        return root

    def buildTree(self, arr, operation):
        self.operation = operation
        self.root = self._buildTree(arr, 0, len(arr)-1)
        
        
    def _getMax(self, node, start, end):
        # Check for Out of Bound
        if end < node.left or node.right < start: 
            if self.operation=='min':return float('inf')
            if self.operation=='max':return float('-inf')
            if self.operation=='sum':return 0
        
        # If this is the node
        if node.left == start and node.right == end: return node.val
        # If node found
        if node.left == node.right: return node.val
        # Overlapping condition
        left  = self._getMax(node.left_child, start, end)
        right = self._getMax(node.right_child, start, end)
        return eval(self.operation)([left, right])        
    
    def query(self, start, end):
        return self._getMax(self.root, start, end)
    
    
    def _update(self, node, index, new_val):
        #print(node.left, node.right, node.val, end=", ")
        if node is None: return None
        if node.left==index and node.right==index:
            temp, node.val  = node.val, new_val
            return temp
        
        mid = (node.left+node.right)//2
        updated_val = self._update(node.left_child, index, new_val) if index<=mid else self._update(node.right_child, index, new_val) 
        
        if self.operation=='sum':
            node.val = node.val - updated_val + new_val
        else:
            node.val = eval(self.operation)([node.val - updated_val, new_val])
        return node.val
    
    def update(self, index, new_val):
        self._update(self.root, index, new_val)
```

```python
arr = [3,4,2,1,6,8,9]
obj = Sum()
obj.buildTree(arr, 'sum')
print(obj.query(0,1))
obj.update(0,100)
print(obj.query(0,1))
```
