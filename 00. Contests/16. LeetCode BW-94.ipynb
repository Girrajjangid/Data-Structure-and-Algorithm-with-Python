```python
import itertools
from functools import reduce
from functools import cache

class Solution:
    def countAnagrams(self, s: str) -> int:

        @cache
        def factorial(n):
           return 1 if n<=0 else n*factorial(n-1)

        def count_unique_permutation(word) :
            max_chart = 26
            length = len(word)
            freq = [0] * max_chart
                
            for i in range(0, length) :
                if (word[i] >= 'a') :
                    freq[(ord)(word[i]) - 97] = freq[(ord)(word[i]) - 97] + 1
            fact = 1
            for i in range(0, max_chart) :
                fact = fact * factorial(freq[i])   
            return factorial(length) // fact
        res = [count_unique_permutation(word) for word in s.split(' ')]
        res = [i for i in res if i!=0]
        return reduce(lambda a,b:a*b, res)%1000000007 if res else 0
```
